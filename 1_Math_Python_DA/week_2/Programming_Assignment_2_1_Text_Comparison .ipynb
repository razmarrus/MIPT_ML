```python
import re
import numpy as np
from collections import Counter
from scipy.spatial import distance
import pandas as pd
```

```python
f = open("sentences.txt", "r")
text = f.read()
```

## 2
Произведите токенизацию, то есть разбиение текстов на слова. Для этого можно воспользоваться регулярным выражением, которое считает разделителем любой символ, не являющийся буквой: re.split('[^a-z]', t). Не забудьте удалить пустые слова после разделения.

```python
def split_into_words(text):
    words = re.split('[^a-z\']', text.lower()) 
    words = [x.lower() for x in words if x]
    return words
```

## 3
Составьте список всех слов, встречающихся в предложениях. Сопоставьте каждому слову индекс от нуля до (d - 1), где d — число различных слов в предложениях. Для этого удобно воспользоваться структурой dict.


```python
sentences = re.split('\n', text)
```

```python
#split_regex = re.compile(r'[.|!|?|…]')
#sentences = filter(lambda t: t, [t.strip() for t in split_regex.split(text)])
```

```python
tracking = []
for sentence in sentences:
    words = split_into_words(sentence)
    words_dict = {}

    for word in words: 
        if word in words_dict.keys():#words_dict.get(word):
            words_dict.update({word: words_dict.get(word)+1})
        else:
            words_dict.update({word: 1})
    tracking.append(words_dict)
#tracking
```

## 4
Создайте матрицу размера n * d, где n — число предложений. Заполните ее: элемент с индексом (i, j) в этой матрице должен быть равен количеству вхождений j-го слова в i-е предложение. У вас должна получиться матрица размера 22 * 254.


```python
print(len(sentences))
```

```python
len(Counter(text.split()))
```

```python
all_words_list = split_into_words(text)
words_list = []
for word in all_words_list:
    if word not in words_list:
        words_list.append(word)
        
len(words_list)
```

```python
df = pd.DataFrame(index=range(1,len(sentences)+1),columns=words_list, data =0)
df
```

```python
for i in range(len(tracking)):
    sentence_dict = tracking[i]
    #print(sentence_dict)
    for key, value in sentence_dict.items():
        for column in df.columns:
            if column == key:
                df.loc[i+1, key] = value
df
```

## 5
Найдите косинусное расстояние от предложения в самой первой строке (In comparison to dogs, cats have not undergone...) до всех остальных с помощью функции scipy.spatial.distance.cosine. Какие номера у двух предложений, ближайших к нему по этому расстоянию (строки нумеруются с нуля)? Эти два числа и будут ответами на задание. Само предложение (In comparison to dogs, cats have not undergone... ) имеет индекс 0.


```python
df.shape
```

```python
scores =[distance.cosine(df.iloc[0], df.iloc[i]) for i in range(1,df.shape[0])]
scores
```

```python
def second_largest(numbers):
    m1, m2 = 0, 0
    index_1, index_2 = None, None
    for i in range(len(numbers)):
        if numbers[i] >= m1:
            m1, m2 = numbers[i], m1
            index_1, index_2 = i, index_1
        elif numbers[i] > m2:
            m2 = numbers[i]
            index_2 = i
    return index_1 + 1, index_2 + 1, m1, m2
```

```python
def second_smollest(numbers):
    m1, m2 = 1, 1
    index_1, index_2 = None, None
    for i in range(len(numbers)):
        if numbers[i] <= m1:
            m1, m2 = numbers[i], m1
            index_1, index_2 = i, index_1
        elif numbers[i] < m2:
            m2 = numbers[i]
            index_2 = i
    return index_1 + 1, index_2 + 1, m1, m2
```

```python
second_smollest(scores)
```
